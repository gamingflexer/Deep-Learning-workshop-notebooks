# Why Deep Learning ?

Why-Deep-Learning-1024x742.webp

Deep learning is a subset of machine learning that is based on the architecture of artificial neural networks. It involves the use of multi-layer neural networks with large numbers of hidden layers, which enable the network to learn complex representations of data.

Deep learning algorithms are able to automatically learn hierarchical representations of features from raw data, such as images, audio, and text, by using multiple layers of non-linear transformations. This allows the algorithm to extract more abstract and complex features at each successive layer, ultimately leading to high-level representations that can be used for tasks such as object recognition, speech recognition, and natural language processing.

One of the key advantages of deep learning is its ability to perform end-to-end learning, where the input data is directly mapped to the output, without the need for intermediate feature extraction. This can lead to improved accuracy and efficiency in many applications.

Deep learning has been used to achieve state-of-the-art performance on a wide range of tasks, including `image classification, object detection, speech recognition, language translation, and game playing,` among others. It has been particularly successful in domains where large amounts of data are available, such as `computer vision and natural language processing`, and has been used to achieve breakthroughs in areas such as self-driving cars and medical diagnosis.

## Perceptron

image.png

Perceptron is a type of artificial neural network that is designed to perform binary classification tasks. It was introduced in the 1950s by psychologist Frank Rosenblatt as a simplified model of biological neurons and has since become a popular algorithm in the field of machine learning.

A perceptron consists of a single layer of artificial neurons, each of which receives one or more input signals and produces a binary output of either 0 or 1. The output of the perceptron is determined by a weighted sum of the inputs, which is passed through an activation function, typically a step function or a sigmoid function.

During training, the weights of the connections between neurons are adjusted based on the error between the predicted output and the actual output. This is done using a supervised learning approach called the perceptron learning rule, which updates the weights in proportion to the input signals and the error signal.

The perceptron algorithm is well-suited for solving linearly separable binary classification problems, where the data points can be divided into two classes by a hyperplane. However, it may not be effective for more complex problems that require non-linear decision boundaries.

Perceptrons have been used in a variety of applications, including image recognition, speech recognition, and natural language processing, among others. They have also been used as building blocks for more complex neural networks, such as multilayer perceptrons and convolutional neural networks.

# Artificial Neural Network(ANN) and its architecture

image.png

`Artificial neural networks (ANNs)` are a type of machine learning algorithm inspired by the structure and function of biological neurons in the human brain. ANNs consist of multiple interconnected layers of artificial neurons, which process information and learn to make predictions or decisions based on input data.

The basic building block of an ANN is an artificial neuron, also called a node or unit, which receives one or more inputs, performs a computation on them, and produces an output. The output of one neuron is then used as input to other neurons in the next layer of the network, until the final output is produced.

`ANNs` are typically trained using a supervised learning approach, where the network is shown a set of labeled training data, and it learns to make predictions by adjusting the weights and biases of the connections between neurons to minimize the error between its predicted output and the actual output. This process is known as backpropagation.

`ANNs` can be used for a wide range of applications, including image recognition, natural language processing, and speech recognition, among others. They have been particularly successful in solving problems that are difficult to solve using traditional programming techniques or require processing large amounts of data.

**`Forward Propagation:`**
Forward propagation is the process by which data is passed through a neural network, layer by layer, to produce an output. During this process, the input data is fed into the first layer of the network, which processes it and passes it on to the next layer. Each layer uses a set of weights and biases to transform the input data in a way that helps the network make predictions. The output of the final layer is the predicted output of the network.

# Activation Functions

image.png

In artificial neural networks, activation functions are mathematical functions that determine the output of a neuron based on its inputs. They introduce non-linearities into the network, allowing it to learn complex representations of data.

There are several common activation functions used in neural networks, including:

`Sigmoid function`: The sigmoid function produces a sigmoid-shaped curve, which is commonly used in binary classification problems. It maps any input to a value between 0 and 1, which can be interpreted as a probability.

`ReLU (Rectified Linear Unit) function`: The ReLU function is a popular choice for deep neural networks. It outputs the input value if it is positive, and 0 if it is negative. This function is computationally efficient and has been shown to improve training times and accuracy in some cases.

`Tanh (Hyperbolic Tangent) function`: The tanh function is similar to the sigmoid function but produces output values between -1 and 1. It is commonly used in recurrent neural networks (RNNs) for sequence modeling tasks.

`Softmax function`: The softmax function is used to produce a probability distribution over multiple classes in multi-class classification problems. It ensures that the sum of the probabilities for all classes adds up to 1.

`Swish function`: The Swish function is a recently introduced activation function which is similar to ReLU, but has a smoother curve which allows it to learn more efficiently in some cases.

The choice of activation function can have a significant impact on the performance of the neural network, and the choice often depends on the specific task and the structure of the network.

In neural networks, ***optimization refers to the process of updating the weights and biases of the network to minimize the error between the predicted output and the actual output for a given set of input data***. The backpropagation algorithm is a popular optimization algorithm used for training neural networks.

Backpropagation works by propagating the error gradient backwards through the network, from the output layer to the input layer. The error gradient represents the rate of change of the network's error with respect to its weights and biases. The weights and biases are then adjusted using this gradient, in the direction that reduces the error, using a method called gradient descent.

There are several variations of gradient descent that can be used in backpropagation, including:

Batch gradient descent: This method updates the weights and biases based on the average error gradient over the entire training dataset.

Stochastic gradient descent (SGD): This method updates the weights and biases based on the error gradient for a single training example at a time. SGD can be faster than batch gradient descent, but can be more noisy and less stable.

Mini-batch gradient descent: This method updates the weights and biases based on the average error gradient over a small subset of the training dataset, called a mini-batch.

In addition to gradient descent, there are other optimization algorithms that can be used in backpropagation, such as momentum, Adagrad, and Adam, which can help improve the speed and stability of the optimization process.

The choice of optimization algorithm can have a significant impact on the training process and the performance of the network, and often depends on the specific task and the structure of the network..  

# Tensors

In mathematics and computer science, a tensor is a multi-dimensional array of numerical values that can be used to represent complex data structures and transformations. Tensors can be thought of as a generalization of scalars, vectors, and matrices to higher dimensions.

In machine learning and deep learning, tensors are a fundamental data structure used to represent and manipulate multi-dimensional data, such as images, audio, and text. In these applications, tensors are typically represented as multi-dimensional arrays with a fixed number of dimensions and a fixed size along each dimension.

For example, a grayscale image can be represented as a 2D tensor with two dimensions - width and height - and each element of the tensor represents the pixel value at that location in the image. Similarly, a color image can be represented as a 3D tensor with three dimensions - width, height, and color channels (red, green, and blue).

Tensors can be manipulated using a variety of operations, such as element-wise operations (e.g., addition, multiplication), matrix operations (e.g., dot product, transpose), and tensor operations (e.g., tensor contraction, tensor product). These operations can be used to perform tasks such as data preprocessing, feature extraction, and model training and evaluation in machine learning and deep learning.

# Sigmoid Activation Function

```python
def sigmoid(z):
    s = 1 / (1 + np.exp(-z))
    return s
```

# Initialization of hyperparameters

```python
def initialize_with_zeros(dim):
    w = np.zeros((dim,1))
    b = 0
    assert(w.shape == (dim, 1))
    assert(isinstance(b, float) or isinstance(b, int))
    return w, b

dim = 2
w, b = initialize_with_zeros(dim)
print ("w = " + str(w))
print ("b = " + str(b))

```

# Forward Propogation

image.png

```python
def propagate(w, b, X, Y):

    m = X.shape[1]
    A = sigmoid(np.dot(w.T, X) + b)
    cost = (- 1 / m) * np.sum(Y * np.log(A) + (1 - Y) * (np.log(1 - A)))

    dw = (1 / m) * np.dot(X, (A - Y).T)
    db = (1 / m) * np.sum(A - Y)

    assert(dw.shape == w.shape)
    assert(db.dtype == float)
    cost = np.squeeze(cost)
    assert(cost.shape == ())
    grads = {"dw": dw,
    "db": db}
    return grads, cost

w, b, X, Y = np.array([[1.],[2.]]), 2., np.array([[1.,2.,-1.],[3.,4.,-3.2]]), np.array([[1,0,1]])
grads, cost = propagate(w, b, X, Y)
print ("dw = " + str(grads["dw"]))
print ("db = " + str(grads["db"]))
print ("cost = " + str(cost))
```

# Backward Propogation and Optimization

```python
def optimize(w, b, X, Y, num_iterations, learning_rate, print_cost = False):
    costs = []
    for i in range(num_iterations):
      grads, cost = propagate(w, b, X, Y)
      dw = grads["dw"]
      db = grads["db"]

      w = w - learning_rate * dw #updating the weights 
      b = b - learning_rate * db #updating the bias
      if i % 100 == 0:
        costs.append(cost)
      if print_cost and i % 100 == 0:
        print("Cost after iteration %i: %f" %(i, cost))
    params = {"w": w,"b": b}
    grads = {"dw": dw,"db": db}
    return params, grads, costs
```

# Prediction Function

```python
def predict(w, b, X):
    m = X.shape[1]
    Y_prediction = np.zeros((1,m))
    w = w.reshape(X.shape[0], 1)
    A = sigmoid(np.dot(w.T, X) + b)
#### VECTORISED IMPLEMENTATION ####
    Y_prediction = (A >= 0.5) * 1.0
    assert(Y_prediction.shape == (1, m))
    return Y_prediction

w = np.array([[0.1124579],[0.23106775]])
b = -0.3
X = np.array([[1.,-1.1,-3.2],[1.2,2.,0.1]])
print ("predictions = " + str(predict(w, b, X)))
```

# Complete Model

```python
def model(X_train, Y_train, X_test, Y_test, num_iterations = 2000, learning_rate = 0.5, print_cost = False):

    w, b = initialize_with_zeros(X_train.shape[0]) # parameters initialization

    parameters, grads, costs = optimize(w, b, X_train, Y_train, num_iterations, learning_rate, print_cost) #forward + backward propogation
    w = parameters["w"] #Equating the weights to optimized weights 
    b = parameters["b"] #equating the bias to optimized bias
    Y_prediction_test = predict(w, b, X_test) #predictions on test set
    Y_prediction_train = predict(w, b, X_train) #prediction on train set

    print("train accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_train - Y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_test - Y_test)) * 100))

    d = {"costs": costs,
    "Y_prediction_test": Y_prediction_test,
    "Y_prediction_train" : Y_prediction_train,
    "w" : w,
    "b" : b,
    "learning_rate" : learning_rate,
    "num_iterations": num_iterations}
    return d
```

# Traning the model

```python
d = model(train_set_x, train_set_y, test_set_x, test_set_y, num_iterations = 1000, learning_rate = 0.005, print_cost = True)
```
