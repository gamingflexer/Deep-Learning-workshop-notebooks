```python
# Import (We are using Tensorflow as backend)
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Checking if i have GPU Enabled
tf.config.list_physical_devices('GPU')
```

```python
# Testing if the col;abs runtime is at GPU # True means yes here
tf.test.is_gpu_available()
```

```python
"""
Here we increase our dataset as we have less data. This is also refered to as Augmentation of images.

Data augmentation is a technique used in deep learning to artificially increase the size of a training dataset by applying various transformations to the existing images. 
The goal of data augmentation is to improve the performance of a deep learning model by providing it with more varied and diverse training data, 
which can help to prevent overfitting and improve the model's ability to generalize to new and unseen data.

1. Train Dataset
2. Test Dataset

"""

training_datagenarator= ImageDataGenerator(rescale=1./255,horizontal_flip=True,
    vertical_flip=True,shear_range=0.2,
    zoom_range=0.2,width_shift_range=0.2,
    height_shift_range=0.2,validation_split=0.1)

test_datagen= ImageDataGenerator(rescale=1./255,horizontal_flip=True,
    vertical_flip=True,shear_range=0.2,
    zoom_range=0.2,width_shift_range=0.2)
```

```python
batch_size = 8
```

```python
# dividing the data into training and validation
train=training_datagenarator.flow_from_directory('/content/drive/MyDrive/#temp/session_dataset/train', target_size=(224, 224),color_mode='rgb', class_mode='categorical', batch_size=batch_size,subset='training')

validation=training_datagenarator.flow_from_directory('/content/drive/MyDrive/#temp/session_dataset/train', target_size=(224, 224),color_mode='rgb', class_mode='categorical', batch_size=batch_size,subset='validation')

#test = test_datagen.flow_from_directory('/content/gdrive/MyDrive/DisasterModel/test', target_size=(224, 224),color_mode='rgb', class_mode='categorical', batch_size=batch_size,shuffle=False)
```

```python
# Imports
from tensorflow.keras.layers import Dense, Activation, Flatten, Dropout, BatchNormalization
```

```python
# Image --> Array of RGB color pixels --> np.array () type matrix  # matrix.size
```

```python
# Initializing CNN
cnn=tf.keras.models.Sequential()

# adding first layer
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,padding='same',activation='relu',input_shape=[224,224,3]))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2))
cnn.add(Dropout(0.25))

# adding second layer
cnn.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,padding='same',activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2))
cnn.add(Dropout(0.25))

cnn.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,padding='same',activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2))
cnn.add(Dropout(0.25))

cnn.add(tf.keras.layers.Conv2D(filters=128,kernel_size=3,padding='same',activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2))
cnn.add(Dropout(0.25))

# adding third layer Extra layer added
cnn.add(tf.keras.layers.Conv2D(filters=256,kernel_size=3,padding='same',activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=1))
cnn.add(Dropout(0.25))

# Flattening
cnn.add(tf.keras.layers.Flatten())
cnn.add(Dropout(0.5))
# Fully connected layer
cnn.add(tf.keras.layers.Dense(units=512,activation='relu'))
cnn.add(BatchNormalization())
# Output layers
cnn.add(tf.keras.layers.Dense(units=2,activation='softmax'))
```

```python
cnn.summary()
```

```python
from keras.callbacks import EarlyStopping 

# After every epoch the models is saved (Weights which get updated) locally and after each iteration of the new models performs better than the old the old "checkpoint is deleted"

checkpoint=tf.keras.callbacks.ModelCheckpoint('./model.h5',
                                           monitor='val_loss',mode="min",save_best_only = True)
```

```python
callbacks=checkpoint
```

```python
# We need to compile the model before starting to train
cnn.compile(optimizer='Adam',loss='categorical_crossentropy',metrics=['accuracy'])

# Here the ".fit" method is used to fit (train) the model on the given dataset.
cnn.fit(train,validation_data=validation,epochs=100,
                  steps_per_epoch=train.samples//batch_size, # steps per epoch parameter is used during training to specify how many batches of samples should be processed per epoch in train
                  validation_steps=validation.samples//batch_size, # same as above in test
                  callbacks=callbacks)
```

```python
from tensorflow.keras.models import load_model

# load_model is the method to load the model from locally as the models are saved in .h5 format (There are some standard formats in which models are stored)
cnn=load_model('/content/model.h5')
```

```python
# Imports
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import PIL
```

```python
# Predictions code
def pred(image_to__open):
    test_image=image.load_img(image_to__open,target_size=(224,224))
    test_image=image.img_to_array(test_image)
    test_image=test_image/255
    test_image=np.expand_dims(test_image,axis=0)
    result=cnn.predict(test_image)

    Catagories=['disaster','non-disaster'] # write all the categories in which binary classfication is done

    image_show=PIL.Image.open(image_to__open)
    plt.imshow(image_show)

    plt.title(Catagories[int(np.argmax(result))])
    plt.show()
```

```python
base_path_images = "/content/05_01_0006.png"

pred(base_path_images)
```
